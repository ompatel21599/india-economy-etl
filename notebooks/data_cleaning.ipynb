```python
import pandas as pd
import numpy as np
import os

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv("../data/india_economy_raw.csv")
print("Shape of data:", df.shape)
print("\nFirst 10 rows:")
df.head(10)
```

```python
print("Missing values in each column:")
print(df.isnull().sum())

print("\nData types:")
print(df.dtypes)

print("\nBasic statistics:")
df.describe()
```

```python
df_pivot = df.pivot_table(
    index='year',
    columns='indicator',
    values='value'
).reset_index()

df_pivot.columns.name = None
df_pivot = df_pivot.sort_values('year', ascending=True).reset_index(drop=True)

print("Pivoted data shape:", df_pivot.shape)
print("\nColumns:", df_pivot.columns.tolist())
df_pivot.head(10)
```

```python
df_clean = df_pivot.dropna()

print("Shape after removing missing years:", df_clean.shape)
print("\nYear range:", df_clean['year'].min(), "to", df_clean['year'].max())
print("\nMissing values:")
print(df_clean.isnull().sum())

df_clean.head(10)
```

```python
print("How many valid values per indicator:")
print(df_pivot.isnull().sum())
print("\nTotal rows:", len(df_pivot))
```

```python
df_clean = df_pivot.drop(columns=['government_debt', 'unemployment_rate'])

df_clean = df_clean.dropna()

print("Shape after cleaning:", df_clean.shape)
print("\nYear range:", df_clean['year'].min(), "to", df_clean['year'].max())
print("\nMissing values:")
print(df_clean.isnull().sum())

df_clean.head(10)
```

```python
os.makedirs("../data", exist_ok=True)
df_clean.to_csv("../data/india_economy_clean.csv", index=False)
print("Clean data saved successfully!")
print("Total rows saved:", len(df_clean))
```
